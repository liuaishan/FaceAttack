```python
import torch
from torch.utils.data import DataLoader
import torchvision.transforms as transforms
import numpy as np
import os
from torch.autograd import Variable
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
import sys
import matplotlib.pyplot as plt 
import matplotlib.image as mpimg
from PIL import Image
```

```python
root_img='DRIVE/test/images/'
root_test ='DRIVE/test/1/'
root = 'DRIVE/test/'
drive_dataset=open('DRIVE/drive.txt')
trainset=[]
for lines in drive_dataset:
    lines = lines.rstrip()       #删除本行末尾的空格
    words = lines.split()
    trainset.append((words[0],words[1]))
    
class drive(torch.utils.data.Dataset): #创建自己的类：CelebA,继承torch.utils.data.Dataset
    def __init__(self,root,dataset,transform=None): #初始化一些需要传入的参数
        self.imgs = dataset
        self.transform = transform
 
    def __getitem__(self, index):    #必须要有，用于按照索引读取每个元素的具体内容，训练时是每个batch的内容
        fn, label = self.imgs[index] #fn是图片path #fn和label分别获得imgs[index]也即是刚才每行中word[0]和word[1]的信息
        img1 = Image.open(root_img+fn)  #按照path读入图片from PIL import Image # 按照路径读取图片
       # data = img1.getdata()
       # data = np.matrix(data,dtype='float')
       # data = np.reshape(data,(584,565))/255
       # img1 = data
       # for i in range(584):
       #     for j in range(564):
       #         if (img1[i,j] - 1 >0.00000001):
       #             img1[i,j] = 1;  
       #         if (img1[i,j] - 0 <0.00000001):
        #            img1[i,j] = 0;
       # img1 = Image.fromarray(data.astype(np.uint8))
        plt.imshow(img1)
        img2 = Image.open(root_test+label).convert('1')
        plt.figure()
        plt.imshow(img2)
       # img2.save('cacaca.gif')
     #   plt.imshow(img2)
        if self.transform is not None:
            img1 = self.transform(img1)  #是否进行transform
            img2 = self.transform(img2)
        return img1,img2  #return回哪些内容，那么我们在训练时循环读取每个batch时，就能获得哪些内容
 
    def __len__(self):  #必须要写，返回的是数据集的长度，也就是多少张图片，要和loader的长度作区分
        return len(self.imgs)
transform = transforms.Compose([transforms.Resize((564,564)),transforms.ToTensor()])
transform 
train_data=drive(root,dataset=trainset,transform=transform)
train_loader=torch.utils.data.DataLoader(dataset=train_data, batch_size=1, shuffle=True)
```

```python

lr_init = 0.001
max_epoch = 10


class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.conv1 = nn.Conv2d(in_channels=1, out_channels=32, kernel_size=3, groups=1,padding=1)
        self.conv2 = nn.Conv2d(in_channels=32, out_channels=32, kernel_size=3, groups=1,padding=1)
        self.pool1 = nn.MaxPool2d(2, 2)  #完事后是一半大小
        self.conv3 = nn.Conv2d(in_channels=32, out_channels=64, kernel_size=(3,3), groups=1,padding=(1,1))
        self.conv4 = nn.Conv2d(in_channels=64, out_channels=64, kernel_size=(3,3), groups=1,padding=(1,1))
        self.up1   = nn.Upsample(scale_factor=2,mode='bilinear')#恢复大小
        self.conv5 = nn.Conv2d(in_channels=64, out_channels=32, kernel_size=(3,3), groups=1,padding=(1,1))
        self.conv6 = nn.Conv2d(in_channels=32, out_channels=1, kernel_size=(3,3), groups=1,padding=(1,1))
        self.initialize_weights()
        self.relu = nn.ReLU(inplace = True)
    def forward(self, x):
        x=self.conv1(x)
        x=self.relu(x)
        #x=nn.Dropout(0.4)
        x=self.conv2(x)
        x=self.relu(x)
        x=self.pool1(x)
        x=self.conv3(x)
        x=self.relu(x)
        x=self.conv4(x)
        x=self.relu(x)
        x=self.up1(x)
        x=self.conv5(x)
        x=self.relu(x)
        x=self.conv6(x)
        print("OKK")
        #x = F.relu(self.conv2(F.dropout(F.relu(self.conv1(x)),p=0.7,training=self.training)))
   #     x = F.relu(self.conv3(self.pool1(F.dropout(x,p=0.7,training=self.training))))
    #    x = F.relu(self.conv4(F.dropout(x,p=0.7,training=self.training)))
    #    x = self.up1(x)
     #   x = F.relu(self.conv5(x))
      #  x = self.conv6(F.dropout(x,p=0.7,training=self.training))
    #    print("OKK")
        return x

    # 定义权值初始化
    def initialize_weights(self):
        for m in self.modules():
            if isinstance(m, nn.Conv2d):
                torch.nn.init.xavier_normal_(m.weight.data)
                if m.bias is not None:
                    m.bias.data.zero_()
            elif isinstance(m, nn.BatchNorm2d):
                m.weight.data.fill_(0)
                m.bias.data.zero_()
            elif isinstance(m, nn.Linear):
                torch.nn.init.normal_(m.weight.data, 0, 0.01)
                m.bias.data.zero_()


net = Net()     # 创建一个网络
net.initialize_weights()
```

```python
criterion = nn.BCEWithLogitsLoss()
optimizer = optim.RMSprop(net.parameters(), lr=lr_init, momentum=0.7) 
#optimizer = optim.Adam(net.parameters(), lr=lr_init)
# 选择优化器
scheduler = optim.lr_scheduler.StepLR(optimizer, step_size=1, gamma=1) 
# 设置学习率下降策略
```

```python
for epoch in range(max_epoch):
    
    loss_sigma = 0.0
    correct = 0.0
    total = 0.0
    scheduler.step()
    #-----------------------加载数据--------------------------------
    for i,data in enumerate(train_loader):
        inputs, labels = data 
        inputs = inputs.float()
        labels = labels.float()
        inputs, labels = Variable(inputs), Variable(labels)
        
        optimizer.zero_grad()
        outputs = net(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        
        _, predicted = torch.max(outputs.data, 1)
        total += labels.size(2)*labels.size(3)
        correct += (predicted == labels.long()).squeeze().sum().numpy()
        loss_sigma += loss.item()
        
        if i%1 == 0:
            loss_avg = loss_sigma
            loss_sigma = 0.0
            print("Training: Epoch[{:0>3}/{:0>3}] Iteration[{:0>3}/{:0>3}] Loss: {:.4f}".format(
                epoch + 1, max_epoch, i + 1, len(train_loader), loss_avg))
```
